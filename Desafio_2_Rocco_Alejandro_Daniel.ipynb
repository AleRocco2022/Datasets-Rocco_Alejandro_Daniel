# Gr√°ficos con Matplotlib.

```python
fig, ax = plt.subplots(figsize=(12,8))
Fecha = df1["last_review"]
Lugar = df1["neighbourhood_group"]
ax.scatter(Fecha, Lugar, alpha=0.7)
ax.set_title("Lugares con mayor inter√©s para alquilar")
ax.set_xlabel("Fecha")
ax.set_ylabel("Lugar")
plt.show()
```

De este primer grafico podemos apreciar que la mayor√≠a de la gente se ubica dentro de New York, en los barrios de Manhattan y Brooklyn, dentro de un per√≠odo de tiempo de an√°lisis que va desde el a√±o 2011 al 2019.

```python
fig, ax = plt.subplots(figsize=(15,4))
Tipo_de_lugar = df1.groupby('room_type')['calculated_host_listings_count'].count()
ax.bar(Tipo_de_lugar.index, Tipo_de_lugar.values)
ax.set_title("Cantidad de Lugares")
ax.set_ylabel("Recuento de Veces")
ax.set_xlabel("Tipo de Lugar")
plt.show()
```

De este segundo gr√°fico podemos observar que la mayor√≠a de las personas que rentan un lugar en New York, lo hacen en departamentos o casas enteras.

```python
fig, ax = plt.subplots(figsize=(15,4))
Tipo_de_lugar = df1.groupby('room_type')['price'].mean()
ax.bar(Tipo_de_lugar.index, Tipo_de_lugar.values, alpha=0.7)
ax.set_title("Precios acorde a la habitaci√≥n")
ax.set_ylabel("Precio promedio")
ax.set_xlabel("Tipo de Lugar")
plt.show()
```

# Gr√°ficos con Seaborn

```python
plt.figure(figsize=(12,7))

type1valores= df1.groupby('room_type').room_type.count()

plt.title("Cantidad de reservas por tipo")

g = sns.lineplot(x=type1valores.index, y=type1valores, )
g.set_xlabel('Tipo')
g.set_ylabel('Count')
plt.show()
```

Este gr√°fico muestra la cantidad de reservas por tipo de habitaci√≥n en un formato lineal, con cada tipo de habitaci√≥n en el eje x y la cantidad de reservas correspondiente en el eje y.

```python
# Crea una nueva columna que represente la cantidad de d√≠as reservados
df1['dias_reservados'] = (df1['last_review'] - df1['last_review']).dt.days

plt.figure(figsize=(12,8))
for room_type in df1['room_type'].unique():
    sns.histplot(df1[df1['room_type'] == room_type]['dias_reservados'], kde=False, label=room_type)
plt.title("Cantidad de d√≠as reservados por tipo")
plt.xlabel('Cantidad de d√≠as reservados')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
# Agrupa por el nombre del host y cuenta las reservas
df_host = df1.groupby('host_name').count()['number_of_reviews'].reset_index()

# Ordena de mayor a menor
df_host = df_host.sort_values(by='number_of_reviews', ascending=False)

# Toma solo los primeros 10
df_host = df_host.head(10)

# Grafica los resultados
plt.figure(figsize=(12,8))
sns.barplot(x='host_name', y='number_of_reviews', data=df_host)
plt.title("Cantidad de veces que el host reserva un tipo de habitaci√≥n")
plt.xlabel('Nombre del host')
plt.ylabel('Cantidad de reservas')
plt.xticks(rotation=90)
plt.show()
```

Se ha agrupado por el nombre del host y contado la cantidad de reservas, y luego ordenamos los resultados de mayor a menor para tomar solo los primeros 10. Finalmente, graficamos estos resultados usando un gr√°fico de barras.

```python
grouped = df1.groupby(['host_name', 'room_type']).count().reset_index()
grouped = grouped[['host_name', 'room_type', 'id']].sort_values(by='id', ascending=False)
grouped = grouped.head(10)

plt.figure(figsize=(12,8))
sns.barplot(x='host_name', y='id', hue='room_type', data=grouped)
plt.title("Cantidad de reservas por host_name y tipo de habitaci√≥n")
plt.xlabel('Host name')
plt.ylabel('Cantidad de reservas')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

Para visualizar la cantidad de reservas por host_name y por tipo de habitaci√≥n, se puede agrupar por ambas columnas y calcular la cantidad de filas en cada grupo. Luego, se puede usar un gr√°fico de barras para visualizar los resultados.

```python
df2 = pd.DataFrame({'last_review': df1['last_review'], 'neighbourhood_group': df1['neighbourhood_group']})

# Convert the 'last_review' column to a datetime format
df2['last_review'] = pd.to_datetime(df2['last_review'])

# Set the figure size
plt.figure(figsize=(15,12))

# Plot the line chart using the seaborn lineplot method
sns.lineplot(x='last_review', y='neighbourhood_group', data=df2)

# Add a title to the chart
plt.title('Line Plot de √∫ltimas reservas en relaci√≥n a la ubicaci√≥n')

# Show the chart
plt.show()
```

Este gr√°fico de l√≠neas nos permite visualizar las cantidad de reservas realizadas a trav√©s del tiempo en los diferentes barrios de New York, por medio de la aplicaci√≥n.

```python
df3 = pd.DataFrame({'neighbourhood_group': df1['neighbourhood_group'],'price': df1['price']})
plt.figure(figsize=(20,15))
sns.boxplot(x='neighbourhood_group', y='price', hue= 'neighbourhood_group', data= df3)
plt.title('Valores de reserva acorde a los distintos lugares')
plt.show()
```

Este boxplot, nos permite visualizar los distintos outlyers en las diferentes zonas, si bien el promedio ronda entre las 140 a 160 d√≥lares, hay muchos casos puntuales en Brooklyn, Manhattan y Queens.
